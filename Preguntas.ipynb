```python
# Preguntas 2
def metodo_gauss_seidel(A,x0,b):
    n = len(b)
    x=x0.copy()
    k = 0
    while k <= 500:
        x_new = x.copy()
        for i in range(n):
            sum1 = np.dot(A[i, :i], x_new[:i])
            sum2 = np.dot(A[i, i + 1:], x[i + 1:])
            x_new[i] = (b[i] - sum1 - sum2) / A[i, i]
        x = x_new
        norma = np.linalg.norm(A@x-b)
        print("El error: ",norma)
        k = k+1
    return x
            
```

```python
def hilbert(k):
    k = int(k)
    H = np.identity(k)
    for i in range(k):
        for j in range(k):
            H[i,j] = 1/(i+j+1)
    return H
```
